```python
import os
import multiprocessing

data = [] 

models = ["maritalk","gpt-4","gpt-3.5-turbo"]
dirr = [
    "../Redações/system prompt vazio/Enem 2022 Sem Texto Motivador/",
    "../Redações/system prompt vazio/Enem 2022 Com Texto Motivador/",
    "../Redações/system prompt matias/Enem 2022 Sem Texto Motivador/",
    "../Redações/system prompt matias/Enem 2022 Com Texto Motivador/"
        ]

directories = [d + m for m in models for d in dirr]
already_processed = multiprocessing.Manager().dict()
```

```python
def process_file(file_path):
    if(file_path in already_processed.keys()):
        return already_processed[file_path]

    with open(file_path, "r") as file:
        print(file_path)
        row = analyze_text(file.read())
        row["type"] = file_path
        already_processed[file_path] = row
        return row

max_threads = multiprocessing.cpu_count() -1  # You can adjust this number based on your system's capacity

# Use ThreadPoolExecutor to parallelize file processing
with multiprocessing.Pool(processes=max_threads) as pool:
    for directory in directories:
        files = [os.path.join(directory, filename) for filename in os.listdir(directory)[:50]]
        results = pool.map(process_file, files)
        data.extend(results)
```

```python
machine = pd.DataFrame(data)
machine["type"] = machine["type"].apply(lambda x: "/".join(x.split("/")[2:-1]))
machine = machine.groupby("type").apply(lambda x: x.sample(25)).reset_index(drop=True)
```

```python
machine.to_csv("data/machines_full.csv",index=False)
```
